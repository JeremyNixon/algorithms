```python
array = [2,8,7,1,3,5,6,4]
print array
```

```python
ar = array
```

```python
def partition(ar):
    pivot = ar[-1]
    i = -1
    for j in range(len(ar)-1):
        if ar[j] < pivot:
            i += 1
            temp = ar[i]
            ar[i] = ar[j]
            ar[j] = temp
    temp = ar[i+1]
    ar[i+1] = pivot
    ar[-1] = temp
    print ar
    return i+1
```

```python
def quicksort(array, p = 0, r = len(array)):
    if p < r:
        q = partition(array)
        quicksort(array, p, q-1)
        quicksort(array, q+1, r)
```
