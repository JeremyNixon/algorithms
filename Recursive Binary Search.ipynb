```python
def bsearch(array, element, floor, ceiling):
    mid = (floor + ceiling)//2
    if floor == ceiling:
        if array[floor] == element:
            return floor
        else:
            print -1
    if array[mid] > element:
        return bsearch(array, element, floor, mid-1)
    elif array[mid] < element:
        return bsearch(array, element, mid+1, ceiling)
    else:
        return mid
```

```python
def lsearch(array, element):
    for i in range(0, len(array)-1):
        if array[i] == element:
            return i
        else:
            return -1
```
